### Create a small corpus and query

```python
# Example mini corpus
# In a real application, this would be a much larger dataset, 
# possibly loaded from files or a database.

corpus = [
    "AI detects algae in water",
    "AI predicts pollution in rivers",
    "IoT sensors for algae detection",
    "AI for image recognition",
    "Satellite data for lakes"
]

# Query
query = "AI algae detection"
```

```python
# Step 1: Tokenization
# You need to install rank_bm25 library: pip install rank_bm25
# This imports the BM25 model from the rank_bm25 library.
from rank_bm25 import BM25Okapi 

# Simple whitespace tokenizer for demonstration purposes
tokenized_corpus = [doc.lower().split() for doc in corpus]
tokenized_query = query.lower().split()
```

```python
#Initialize BM25 and get scores
# Step 2: BM25 Scoring 

bm25 = BM25Okapi(tokenized_corpus)
scores = bm25.get_scores(tokenized_query)
```

```python
#   Step 3: Display results
for i, s in enumerate(scores):
    print(f"Doc {i+1} → Score: {s:.3f} | {corpus[i]}")

#Higher score means the document is more relevant.
```

```python
ranked = sorted(
    zip(corpus, scores), # Combine documents with their scores
    key=lambda x: x[1], # Sort by score (x[1]) in descending order
    reverse=True # Set reverse to True for descending order
)

for rank, (doc, score) in enumerate(ranked, start=1):
    print(f"{rank}. {doc}")
    print(f"   Score: {score:.4f}\n")
```

```python
top_docs = bm25.get_top_n( 
    tokenized_query, # The tokenized query
    corpus,
    n=3
)

for doc in top_docs:
    print(doc)
```
