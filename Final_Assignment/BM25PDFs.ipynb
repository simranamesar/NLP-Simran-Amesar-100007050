```python
from pypdf import PdfReader
from rank_bm25 import BM25Okapi
```

```python

#Extract text from PDF

def extract_text_from_pdf(pdf_path):
    reader = PdfReader(pdf_path)

    text = ""

    for page in reader.pages:
        page_text = page.extract_text()

        if page_text:
            text += page_text + " "

    return text
```

```python

pdf_files = [
    "ToyData/learning-langchain-for-true-epub-9781098167288.pdf",
    "ToyData/Natural-Language-Processing-Python.pdf",
    "ToyData/test.pdf"
]


# Treat each PDF as one document 
documents = []

for pdf in pdf_files:
    text = extract_text_from_pdf(pdf)
    documents.append(text)


# Tokenize documents
tokenized_documents = [
    doc.lower().split()
    for doc in documents
]


#  Create BM25 model
#  User query
query = "What is Tokenization?"

tokenized_query = query.lower().split()


# 7. Calculate BM25 scores for each PDF
bm25 = BM25Okapi(tokenized_documents)
scores = bm25.get_scores(tokenized_query)


# Print scores
for pdf, score in zip(pdf_files, scores):
    print(f"{pdf}: {score:.4f}")
```
